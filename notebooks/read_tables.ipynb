```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
conn = sqlite3.connect('../data/td_V1.db')
git_commits = pd.read_sql_query("SELECT * FROM GIT_COMMITS",conn)
szz_fault_inducing_commits = pd.read_sql_query("SELECT * FROM szz_fault_inducing_commits",conn)
refactoring_miner = pd.read_sql_query("SELECT * FROM refactoring_miner",conn)
```

```python
print(git_commits.shape)
git_commits = git_commits[['projectID','commitHash','commitMessage','committerDate','branches']]
git_commits.head()
```

```python
print(szz_fault_inducing_commits.shape)
szz_fault_inducing_commits.head()
```

```python
print(refactoring_miner.shape)
refactoring_miner.head()
```

## JOINS

```python
fault_inducing_commits = szz_fault_inducing_commits.merge(git_commits, left_on = 'faultInducingCommitHash', right_on = 'commitHash')
print(fault_inducing_commits.shape)
fault_inducing_commits.head()
```

```python
fault_fixing_commits = szz_fault_inducing_commits.merge(git_commits, left_on = 'faultFixingCommitHash', right_on = 'commitHash')
print(fault_fixing_commits.shape)
fault_fixing_commits.head()
```

```python
print(fault_fixing_commits.commitHash.unique(), 'unique fault fixingcommits')
```
